#### Add columns to marvels table 

```python
%%sql

ALTER TABLE marvels
    ADD COLUMN durability_level TEXT;
ALTER TABLE marvels
    ADD COLUMN intelligence_level TEXT;
ALTER TABLE marvels
    ADD COLUMN strength_level TEXT;
ALTER TABLE marvels
    ADD COLUMN strength_desc TEXT;
ALTER TABLE marvels
    ADD COLUMN speed_level TEXT;
ALTER TABLE marvels
    ADD COLUMN speed_desc TEXT;
ALTER TABLE marvels
    ADD COLUMN energy_proj_level TEXT;
ALTER TABLE marvels
    ADD COLUMN energy_proj_desc TEXT;
ALTER TABLE marvels
    ADD COLUMN fighting_level TEXT;
```

#### Set the new columns in marvels table to pull data from Power Grid Table

```python
%%sql

UPDATE marvels
SET durability_level = 
    (SELECT durability_level 
        FROM powergrid 
        WHERE attribute_level = marvels.durability);

UPDATE marvels
SET intelligence_level = 
    (SELECT intelligence_level 
        FROM powergrid 
        WHERE attribute_level = marvels.intelligence);

UPDATE marvels
SET strength_level = 
    (SELECT strength_level 
        FROM powergrid 
        WHERE attribute_level = marvels.strength);

UPDATE marvels
SET strength_desc = 
    (SELECT strength_desc 
        FROM powergrid 
        WHERE attribute_level = marvels.strength);

UPDATE marvels
SET speed_level = 
    (SELECT speed_level 
        FROM powergrid 
        WHERE attribute_level = marvels.speed);

UPDATE marvels
SET speed_desc = 
    (SELECT speed_desc 
        FROM powergrid 
        WHERE attribute_level = marvels.speed);

UPDATE marvels
SET energy_proj_level = 
    (SELECT energy_proj_level
        FROM powergrid 
        WHERE attribute_level = marvels.energy_Projection);

UPDATE marvels
SET energy_proj_desc = 
    (SELECT energy_proj_desc 
        FROM powergrid 
        WHERE attribute_level = marvels.energy_Projection);

UPDATE marvels
SET fighting_level = 
    (SELECT fighting_level 
        FROM powergrid 
        WHERE attribute_level = marvels.fighting_Skills);
```

*Using subqueries in this way can simplify the UPDATE queries and make them more efficient by avoiding the need to join large tables and perform additional processing.*

### 3: Answers & Tasks

#### How many characters are in each side (Anti-hero, Bad, & Good)?

```python
%%sql

--Get count of characters in each category
SELECT
    alignment AS Side,
    COUNT(name) AS num_Charactears
    FROM marvels
    GROUP BY alignment;
```

 >**Answer:** *There are 4 villains, 15 heroes, and 7 anti-heroes*

#### What are the average stats for each alignment?

```python
%%sql

SELECT
    alignment,
    ROUND(AVG(intelligence),2) as avg_intel,
    ROUND(AVG(strength),2) as avg_strength,
    ROUND(AVG(speed),2) as avg_speed,
    ROUND(AVG(durability),2) as avg_durability,
    ROUND(AVG(energy_Projection),2) as avg_energy_projection,
    ROUND(AVG(fighting_Skills),2) as avg_fighting_skill
    FROM marvels
    GROUP BY alignment
    ORDER BY avg_intel DESC;
```

> **Answer:** *Villians have a higher average of intelligence, strength, speed, durability, and energy projection;
  Heroes have the best stats for fighting skills, but that might be skewed considering they team up and jump the villians*

#### Grouping Characters by Intelligence and counting by by alignment

```python
%%sql
SELECT
    CASE
        WHEN intelligence >= 5 THEN 'Genius+ IQ'
        WHEN intelligence BETWEEN 3 AND 5 THEN 'Above Normal IQ'
        ELSE 'Normal IQ'
    END as intel_group,
    alignment,
    COUNT(*)
    FROM marvels
    GROUP BY intel_group, alignment
    ORDER BY intel_group
```

> **Answer:** *Despite there being more heroes than villians and anti-heroes combined, both groups have a higher proportion of their
group with 'Genius+ IQs (i.e., 2 anti-heroes, 3 villians, 2 hero)*

#### Who are the characters in the 'HIGH IQ' group

```python
%%sql

SELECT 
    alignment,
    name,
    intelligence_level,
    ROW_NUMBER() OVER (PARTITION BY alignment ORDER BY intelligence DESC) AS rn
    FROM marvels
    WHERE intelligence >= 5;
```

> **Answer:** *Rogue is listed at the top of the list at 7. 
Iron Man (listed as a anti-hero, which makes since given the age of Ultron) is tied w/ Magneto, Thanos, and Doctor Doom at 6.

### While Intelligence is important, but so is durability and strength. You are great have to one or the other.

```python
%%sql

WITH genius_iq_characters AS (
    SELECT 
        alignment,
        name,
        intelligence AS intelligence_level,
        ROW_NUMBER() OVER (PARTITION BY alignment ORDER BY intelligence DESC) AS rn
    FROM marvels
    WHERE intelligence >= 5
)

SELECT *
FROM marvels
WHERE popularity < 25
    AND (durability >= 5 OR strength >= 5)
    AND name IN (
        SELECT name FROM genius_iq_characters
    );
```

```python
%%sql

WITH iq_characters AS (
    
    SELECT 
        name,
        alignment,
        intelligence AS intelligence_level,
        ROW_NUMBER() OVER (PARTITION BY alignment ORDER BY intelligence DESC) AS rn
    FROM marvels
    WHERE intelligence >= 5
)

SELECT 
    g.name,
    m.alignment,
    m.intelligence_level,
    m.strength_level,
    m.strength_desc,
    m.speed_level,
    m.speed_desc
FROM iq_characters g
LEFT JOIN marvels m
    ON g.name = m.name
WHERE m.popularity < 25
    AND (m.durability >= 5 OR m.strength >= 5);
```
